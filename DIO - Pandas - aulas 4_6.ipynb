```
# Importando a biblioteca
import pandas as pd
```

```
# Upload de arquivos - nesse caso ficam temporariamente a disposição
# Leitura de arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```
# Contatenando os arquivos em uma só tabela
# Mostrando as 5 primeiras linhas
df = pd.concat([df1,df2,df3,df4,df5])
df.head()
```

```
# Mostrando as 5 primeiras linhas da tabela unida
df.tail()
```

```
# Trazendo 7 linhas  para exemplificar a tabela toda
df.sample(7)
```

```
# Verificando o tipo de dado de cada coluna
df.dtypes
```

```
# Alterando o tipo de dado da coluna LojaID (pois embora seja um número não serão feitos cálculos com ela)
df["LojaID"] = df["LojaID"].astype("object")
df.dtypes
```

```
# Verificando se há linhas com valores faltantes
df.isnull().sum()
```

```
# Há valores nulos na coluna de vendas - nesse caso há várias abordagens possível:  exclusão das linhas, inserção de valor 0, colocação da média de vendas no lugar
# Colocação da média de vendas
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
df["Vendas"].isnull().sum()
```

```
# Colocação de 0 no lugar de nulo
df["Vendas"].fillna(0, inplace=True)
df["Vendas"].isnull().sum()
```

```
# Apagando linhas com valores nulos
df.dropna(inplace=True)
```

```
# Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

```
# Criando uma nova coluna no dataset = receita = vendas * qtdds
df["Receita"] = df["Vendas"] * df["Qtde"]
df.head()
```

```
# Retornando a maior receita
df["Receita"].max()
```

```
# Retornando as 3 maiores receitas
df.nlargest(3, "Receita")
```

```
# Retornando as 3 menores receitas
df.nsmallest(3, "Receita")
```

```
# Agrupando receita por cidade
df.groupby("Cidade")["Receita"].sum()
```

```
# Ordenando o conj de dados
df.sort_values("Receita", ascending= False).head(10)
```

Trabalhando com datas - Aula 6

```
# transformando a coluna de Data em tipo inteiro
# No cotidiano encontra-se a tabela Data em outro formato que não DateTime
df["Data"] = df["Data"].astype("int64")
```

```
# Verificando o tipo de dado de cada coluna
df.dtypes
```

```
df["Data"] = pd.to_datetime(df["Data"])
df.dtypes
```

```
# Obtendo a receita por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```
# Extraindo somente o ano para criar nova coluna
df["Ano_Venda"] = df["Data"].dt.year
df.sample(3)
```

```
# Extraindo o mês e o dia da venda para criar novas colunas
df["mes_venda"], df["dia_venda"] = df["Data"].dt.month, df["Data"].dt.day
df.sample(6)
```

```
#retornando a data mais antiga
df["Data"].min()
```

```
#retornando a data mais recente
df["Data"].max()
```

```
#calculando a diferença de dias
df["dif_de_dias"] = df["Data"] - df["Data"].min()
df.sample(7)
```

```
#criando a coluna de trimestre
df["trimestre"] = df["Data"].dt.quarter
df.sample(8)
```

```
#filtrando as vendas de 2019 (no mês de março)
vendas_201903 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
vendas_201903.sample(2)
```
